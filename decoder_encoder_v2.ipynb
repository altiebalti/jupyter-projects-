```python
class NN(torch.nn.Module):
    def __init__(self):
        super().__init__()        
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(3 * IMG_SIZE * IMG_SIZE, 256), # (N, 784) -> (N, 128)
            torch.nn.ReLU(),
            torch.nn.Linear(256, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, 12),
            torch.nn.ReLU(),
            torch.nn.Linear(12, 3) # -> N, 3
        )
        
        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(3, 12),
            torch.nn.ReLU(),
            torch.nn.Linear(12, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 256),
            torch.nn.ReLU(),
            torch.nn.Linear(256, IMG_SIZE * IMG_SIZE * 3),
            torch.nn.Sigmoid()
        )
        
    def forward(self, x):
        size1 = x.size(1)
        size2 = x.size(2)
        size3 = x.size(3)
        #print(x.size(0), size1, size2, size3)
        x = x.view(x.size(0), size1*size2*size3)
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        
        return decoded.view(decoded.size(0), size1, size2, size3)

model = NN()
```
